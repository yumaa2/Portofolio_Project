```python
# mengisi missing data

df.budget = df.budget.fillna('0')
```

# Mengubah Tipe Data Kolom

```python
# mengubah tipe data kolom

df['budget'] = df['budget'].astype('int64')
df.info()
```

# Mencocokkan Tahun pada Kolom Rilis dengan Kolom Tahun

```python
# mencocokkan tahun pada kolom rilis dengan kolom tahun

df['released'] = df['released'].astype(str)
df['released']
```

```python
# Membuat kolom year yang benar

df['year_fixed'] = df.released.str.split().str[2]
df
```

## Dengan Scatter Plot

```python
plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Korelasi Budget dan Gross')
plt.xlabel('Budget')
plt.ylabel('Gross')
```

## Regresi

```python
sns.regplot(
    x = df['budget'], 
    y=df['gross'], 
    data=df, 
    scatter_kws={'color':'blue'}, 
    line_kws={'color':'red'})
plt.title('Korelasi Budget dan Gross')
plt.xlabel('Budget')
plt.ylabel('Gross')
```

## Korelasi Data Kategorikal

Numerasi kolom object, seperti kolom company, genre, dsb

```python
df_numerasi = df

for nam_kol in df_numerasi.columns:
    if (df_numerasi[nam_kol].dtype == 'object'):
        df_numerasi[nam_kol] = df_numerasi[nam_kol].astype('category')
        df_numerasi[nam_kol] = df_numerasi[nam_kol].cat.codes  # memberikan numerasi value pada kolom

df_numerasi
```

```python
df 
```

```python
plt.figure(figsize=(16,12))
sns.heatmap(df_numerasi.corr(method='spearman'), annot=True)
```

```python
correlation_mat = df_numerasi.corr(method='spearman')

corr_pairs = correlation_mat.unstack()
pd.DataFrame({'Korelasi':corr_pairs.head(300)})
```

```python
correlation_mat = df_numerasi.corr(method='spearman')

corr_pairs = correlation_mat.unstack().sort_values()
corr_pairs
```
